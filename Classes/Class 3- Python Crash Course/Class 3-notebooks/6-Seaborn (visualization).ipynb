## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Seaborn comes with built-in data sets!
# sns.get_dataset_names()
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
df.info()    
```

## 1. distribution plots

```python
sns.distplot(df['mpg'])
plt.show()
```

```python
sns.distplot(df['mpg'],bins=20, color="red")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg', y='horsepower', data=df, hue='cylinders', size='weight')
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',data=df)
```

```python
sns.jointplot(x='weight',y='mpg',data=df,kind='hex')
```

```python
sns.jointplot(x='weight',y='mpg',data=df ,kind='reg')
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
df.head(5)
```

```python
df1 = df[['mpg','cylinders','weight']]
df1.head(5)
```

```python
sns.pairplot(df1) # pairplot look at the joint relationship between numerical variables in the dataframe. 
plt.show()
```

```python
df1['cylinders'].value_counts()
```

```python
sns.pairplot(df1,hue='cylinders',palette='coolwarm') 
# hue is used for categorical variables
# palette specifies the colors
```

## 2. Categorical plots

```python
sns.boxplot(x='cylinders', y='mpg', data=df)
```

```python
sns.barplot(x='cylinders', y='mpg', data=df)
# barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean.

# the black vertical lines are the error bars. 
```

```python
sns.countplot(x='cylinders',data=df)
```

## 3. Heatmap

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm',annot=True)
plt.show()
```

## 4. Facet grids

```python
df.head()
```

```python
df['origin'].value_counts()
```

```python
g = sns.FacetGrid(df, col="cylinders", row='origin')
g = g.map(plt.scatter, "mpg", "weight") # Apply a plotting function to each facet's subset of the data.
```

```python
g = sns.FacetGrid(df, col="cylinders", hue='origin')
g = g.map(plt.scatter, "mpg", "weight").add_legend()
```

## 5. Regression plots

```python
df.head()
```

```python
sns.lmplot(x='weight',y='mpg',data=df)
```

```python
sns.lmplot(x='weight',y='mpg',data=df, hue='origin', palette='coolwarm')
```

```python
sns.lmplot(x='weight',y='mpg',data=df, col='origin')
```
