```python
from sklearn.metrics import precision_score, recall_score, make_scorer, f1_score

# print(f1_score(y,y_pp))
# print(recall_score(y,y_pp))
# print(precision_score(y,y_pp))

grid = GridSearchCV(
                    estimator = pipeline,
                    param_grid = {'model__class_weight' : [{0:1, 1:x} for x in np.linspace(1,20,30)]},
                    scoring = {'precision' : make_scorer(precision_score),
                               'recall' : make_scorer(recall_score),
#                                'combo' : make_scorer(f1_score)
                               'combo' : make_scorer(min_recall_precision)},
                    refit = 'combo',
                    return_train_score=True,
                    cv = 10,
                    n_jobs = -1                  
                    )
grid.fit(X,y)
pd.DataFrame(grid.cv_results_)
```

```python
plt.figure(figsize=(12,4))
df = pd.DataFrame(grid.cv_results_)
for score in ['mean_test_precision',"mean_test_recall","mean_test_combo"]:
    plt.plot([b[1] for b in df['param_model__class_weight']],df[score],label=score)
plt.legend()
```

```python
plt.figure(figsize=(12,4))
df = pd.DataFrame(grid.cv_results_)
for score in ['mean_train_precision',"mean_train_recall","mean_train_combo"]:
    plt.scatter([_[1] for _ in df['param_model__class_weight']],df[score.replace('test','train')],label=score)
plt.legend()
```
