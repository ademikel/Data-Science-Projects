CREATING MULTIPLE SUBPLOTS

```python
plt.figure(figsize=(10, 4))

figure,a = plt.subplots(2,2)
a[0][0].plot(x,y,'r^-')
a[0][0].set_title('x against y')
a[0][1].plot(y,x,'g*-')
a[0][1].set_title('y against x')
a[1][0].plot(x,y,'b^-')
a[1][0].set_title('x against y')
a[1][1].plot(y,x,'y*-')
a[1][1].set_title('y against x')

plt.show()
```

CREATING A SUBPLOT

This consists of many plot (or subplot in a plot).

plt.subplot(nrows, ncols, plot_number)

nrows is the number or rows i.e. like the example below, i want all the subplots to be on a single row, so i put the nrows as 1.

ncols is the number or columns i.e. like the example below, i want all the subplots to be on a single row, but on several columns so i put the ncols to 4.

plot_number or index is the position each subplot on the plot. so, here we have four plots. Therefore, 1,2,3,4 are the plot_numbers.



```python
# adjusting the size
fig =plt.figure(figsize=(10,10))

# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,4,1)
plt.plot(x, y, 'r') 
plt.subplot(1,4,2)
plt.plot(y, x, 'g')
plt.subplot(1,4,3)
plt.plot(x, y, 'r--')
plt.subplot(1,4,4)
plt.plot(y, x, 'g--')

plt.subplot(4,1,1)
plt.plot(x, y, 'r') 
plt.subplot(4,1,2)
plt.plot(y, x, 'g')
plt.subplot(4,1,3)
plt.plot(x, y, 'r--')
plt.subplot(4,1,4)
plt.plot(y, x, 'g--')

plt.show()
```
