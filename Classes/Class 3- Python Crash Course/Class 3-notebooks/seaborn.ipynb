DISPLOT

```python
sns.distplot(df['mpg'],kde=True,bins=20)
```

Data binning or discrete binning or bucketing involves; 

They divide the range of values int0 series of intervals called bins, and then counts how many values fall into each interval.

x-axis denotes the number of bins, this is to say that when bins = 10, x-axis has 10 bars, which is the same as the number of the bin.
so if bins = 200, x-axis also has 200 (bars) as corresponding to the number of bins.

y-axis represents the number of a particular bin,that is, y-axis contains how much values is contained in each bin.

```python
# bin with specific number in the x-axis
sns.displot(df['mpg'],bins=20,edgecolor=None,color='red')
# kde = false by default
# figure size does not apply to seaborn's charts
# plt.figure(figsize=(40,100))
plt.show()
```

```python
# bin with specific number in the x-axis
sns.displot(df['mpg'],kde=True,bins=20,edgecolor=None,color='orange')
plt.show()
```

```python
# bin within a range
sns.displot(df['mpg'],kde=True,bins=range(0,50,5))
plt.show()
```

```python
# gives all bins(data) available in the dataset
sns.displot(df['mpg'],kde=True,bins=range(0,50))
plt.show()
```

```python
# specifying bin range, maybe i want to analyse a particular part of my dataset, i can specify the range i want
sns.displot(df['mpg'],kde=True,bins=range(20,40,5))
plt.show()
```

```python
# we can control how frequency, that is, if we want all data to be visible on the chart, we can use the 'log=True' which converts the 'count' or frequecy of the data values to logarithm form.
sns.displot(df['mpg'],kde=True,bins=range(0,50), log=True)
# for example, now we know that, we have only 10^0 i.e. 1 number of '9' in the dataset, only one 9 is present in the dataset
plt.show()
```

PAIRPLOT

```python
df.head()
```

```python
mpg = df[(df['mpg']<11)]['cylinders'].to_frame()
mpg
```

```python
# df['origin'].value_counts()
df['origin'].replace({'usa':1,'japan':2,'europe':3},inplace=True)
mpg = df[['mpg','cylinders','weight','origin']]
mpg.head()
sns.pairplot(mpg)
# plt.show()
# In pairplot, correlation plots (e.g scatterplot) are used for heterogenous relation like, mpg and cylinder, weight and cylinder,
# while distribution plots are used for homogenous relationships like mpg and mpg, cylinders and cylinders, and so on. 
```

```python
sns.pairplot(mpg, hue='origin',palette='coolwarm')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
